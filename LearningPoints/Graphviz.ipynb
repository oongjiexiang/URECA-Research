```python
from graphviz import Graph
```

```python
dot = Graph(comment = "My first trial of GraphViz", format='png')      # dot.format = 'svg'
dot.node('A', 'Time Period')
dot.node('B', 'CPI')
dot.node('C', 'Instruction Count')
dot.edges(['AC', 'AB'])
dot.edge('B', 'C', constraint = 'false')
print(dot.source)
```

```python
dot.render('./trial_ACOA_iron_triangle.gv', view=True)       # dot.render()
```
